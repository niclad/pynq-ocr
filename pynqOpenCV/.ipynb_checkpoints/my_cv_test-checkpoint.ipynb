```python
import numpy as np
import time
import cv2
import PIL.Image

in_img = cv2.imread('../a_photo_but_smaller.jpg', flags=0)
print(in_img.shape)
image = PIL.Image.fromarray(in_img)
image
```

```python
# Sobel vertical filter
kernelF = np.array([[1.0,0.0,-1.0],[2.0,0.0,-2.0],[1.0,0.0,-1.0]],np.float32)

start = time.time()
out_img = cv2.filter2D(in_img, -1, kernelF)
end = time.time()
print("FPS: %0.2f" % (1 / (end - start)))
```

```python
from threading import Thread
import pynq_cv.overlays.xv2Filter2DDilate as xv2
import numpy as np
import pynq.buffer as pb

kernel_g = np.array([[1.0,0.0,-1.0],[2.0,0.0,-2.0],[1.0,0.0,-1.0]],np.float32)

np_img = np.array(in_img)
x, y = np_img.shape
xv2_in = in_img

def loop_hw2_app():
    global kernel_g
    
    xv2_out = pb.allocate(in_img.shape, dtype=uint8)
    start=time.time()
    xv2.filter2D(in_img, -1, kernel_g, dst=xv2_out, borderType=cv2.BORDER_CONSTANT)
    end=time.time()
    print("FPS: %0.2f" % (1 / (end - start)))
    
loop_hw2_app()
```
